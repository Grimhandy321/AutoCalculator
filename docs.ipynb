# Predikce ceny ojetých vozidel

Cílem notebooku je vytvořit regresní model, který dokáže predikovat cenu ojetého vozidla na základě reálně sesbíraných dat.

## Dataset
Data byla získána vlastním web scrapingem z aktuálních nabídek autobazarů (www.autojarov.cz).

Dataset obsahuje:
- 1500+ záznamů (2000)
-  atributů:
  - brand
  - fuel
  - transmission
  - km
  - year
  - price

Nebyl použit žádný hotový dataset.
